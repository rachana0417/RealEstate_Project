```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

print("Libraries imported successfully!")
```

```python
df = pd.read_csv("../data/india_housing_prices.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
# Fill numerical missing values with median
df = df.fillna(df.median(numeric_only=True))

# Fill categorical missing values with mode
df = df.fillna(df.mode().iloc[0])

print("Missing values handled!")
```

```python
df.drop_duplicates(inplace=True)
print("Duplicates removed!")
```

```python
df["Price_per_SqFt"] = (df["Price_in_Lakhs"] * 100000) / df["Size_in_SqFt"]
df["Price_per_SqFt"].head()
```

```python
df["Age_of_Property"] = 2025 - df["Year_Built"]
df["Age_of_Property"].head()
```

```python
df["Amenity_Count"] = df["Amenities"].astype(str).apply(lambda x: len(x.split(',')))
df["Amenity_Count"].head()
```

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = ["State", "City", "Locality", "Property_Type", "Furnished_Status", 
            "Security", "Facing", "Owner_Type", "Availability_Status"]

le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))

df[cat_cols].head()
```

```python
median_price = df["Price_per_SqFt"].median()
df["Good_Investment"] = (df["Price_per_SqFt"] < median_price).astype(int)

df["Good_Investment"].head()
```

```python
cat_cols = ["State", "City", "Locality", "Property_Type", "Furnished_Status", 
            "Security", "Facing", "Owner_Type"]

le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))

df[cat_cols].head()
```

```python
df.to_csv("../data/cleaned_data.csv", index=False)
print("Cleaned dataset saved successfully!")
```
